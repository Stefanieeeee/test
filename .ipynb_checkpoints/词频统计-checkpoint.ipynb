```python
##统计频次

for word in words:          # 枚举在文章中出现的词汇
    if len(word) < 1:       # 排除字长小于2的词
        continue
    else:
                count[word] = count.get(word, 0)+1# 统计词频
list2 = list(count.items())     # 将字典中的键值对转化为列表
list2.sort(key=lambda x: x[1], reverse=True)    # 对列表按照词频从大到小排列
```

## 标题和正文情感分析

```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-
import pandas as pd
from pprint import pprint
#from paddlenlp import Taskflow
from datetime import datetime
from tqdm import tqdm
```

```python

_RAW_DATA = "小红书关键词笔记数据_20230807.xlsx"

# 原始评论
df_ori= pd.read_excel(_RAW_DATA)
```

```python

# 原始评论
df = pd.read_excel(_RAW_DATA)

# 评论分析
schema = "情感倾向[正向，负向]"  # Define the schema for sentence-level sentiment classification
ie = Taskflow("sentiment_analysis", schema=schema, batch_size=64)
for i in tqdm(df.index, desc=">>> Analysing sentiment"):
    if df.iloc[i]["笔记标题"]:
        comment = df.iloc[i]["笔记标题"]
        res = ie(str(comment))[0]
        df.loc[i, "笔记标题情感"] = res["label"]
        df.loc[i, "笔记标题情感分数"] = res["score"]
    if df.iloc[i]["笔记正文"]:
        comment = df.iloc[i]["笔记正文"]
        res = ie(str(comment))[0]
        df.loc[i, "笔记正文情感"] = res["label"]
        df.loc[i, "笔记正文情感分数"] = res["score"]

# 保存
# df.to_csv("./result.csv", index=False)
```
